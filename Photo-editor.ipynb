```python
from random import random
import random as rand
from tkinter import *
from tkinter import ttk
from tkinter import filedialog
from tkinter.filedialog import askopenfilename, asksaveasfilename, asksaveasfile
import cv2
import cv2 as cv
import numpy as np
import numpy
import PIL
from PIL import Image, ImageTk, ImageFilter, ImageEnhance, ImageOps
import os
import powerlaw
import math
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import tkinter.messagebox
import urllib.request
import turtle
root = Tk()
#root.state('zoomed')
root.geometry("650x450")

def load():
    global img_path,img
    filename = filedialog.askopenfilename(initialdir="/", title="Select file",filetypes=(("jpeg files", "*.jpg"), ("all files", "*.*")))
    img_path = filename
    try:
        img = Image.open(img_path)
        img1=img
        size = 400, 400
        img.thumbnail(size, Image.ANTIALIAS)
        img1 = ImageTk.PhotoImage(img)
        canvas2.create_image(300, 210, image=img1)
        canvas2.image = img1
    except AttributeError:
        canvas2.create_image(300, 210, image=img1)
        canvas2.image = img1
'''
def save():
    global img_path
    im=Image.open(img_path)
    ext = img_path.split(".")[-1]
    filename = asksaveasfilename(defaultextension=f".{ext}",filetypes=[("All Files", "*.*"), ("PNG file", "*.png"), ("jpg file", "*.jpg")])
    im.save(filename)
'''
img1 = None
img2 = None
img3 = None
img4 = None
img5 = None
img6 = None
img7 = None
img8 = None
img9 = None
img10 = None
img11 = None
img12 = None
img13 = None
img14 = None
img15 = None
img16 = None
img17 = None
img18 = None
img19 = None
img20 = None
img21 = None
img22 = None

def save():
    global img_path, imgg
    # file=None
    ext = img_path.split(".")[-1]
    file = asksaveasfilename(defaultextension=f".{ext}",
                             filetypes=[("All Files", "*.*"), ("PNG file", "*.png"), ("jpg file", "*.jpg")])
    if file:
        cv2.imwrite(file, imgg)
        
def blur(event=None):
    global img_path, img1, imgg
    img = Image.open(img_path)
    img.thumbnail((350, 350))
    imgg = img.filter(ImageFilter.BLUR())
    img1 = ImageTk.PhotoImage(imgg)
    canvas2.create_image(300, 210, image=img1)
    canvas2.image = img1


def boxblur(event=None):
    global img_path, img2, imgg
    img = Image.open(img_path)
    img.thumbnail((350, 350))
    imgg = img.filter(ImageFilter.BoxBlur(5))
    img2 = ImageTk.PhotoImage(imgg)
    canvas2.create_image(300, 210, image=img2)
    canvas2.image = img2


def contour(event=None):
    global img_path, img3, imgg
    img = Image.open(img_path)
    img.thumbnail((350, 350))
    imgg = img.filter(ImageFilter.CONTOUR)
    img3 = ImageTk.PhotoImage(imgg)
    canvas2.create_image(300, 210, image=img3)
    canvas2.image = img3


def emboss(event=None):
    global img_path, img4, imgg
    img = Image.open(img_path)
    img.thumbnail((350, 350))
    imgg = img.filter(ImageFilter.EMBOSS)
    img4 = ImageTk.PhotoImage(imgg)
    canvas2.create_image(300, 210, image=img4)
    canvas2.image = img4


def edgeenhance(event=None):
    global img_path, img5, imgg
    img = Image.open(img_path)
    img.thumbnail((350, 350))
    imgg = img.filter(ImageFilter.EDGE_ENHANCE)
    img5 = ImageTk.PhotoImage(imgg)
    canvas2.create_image(300, 210, image=img5)
    canvas2.image = img5
    
def logtransform():
    global img_path, img6, imgg
    img = Image.open(img_path)
    img.thumbnail((350, 350))
    imgUMat = img
    gray = cv2.cvtColor(np.float32(imgUMat), cv2.COLOR_RGB2GRAY)
    gray = np.uint8(imgUMat)
    c = 255/(np.log(1 + np.max(gray)))
    imgg = c * np.log(1 + gray)
    imgg = np.array(imgg, dtype=np.uint8)
    color_coverted = cv2.cvtColor(imgg, cv2.COLOR_BGR2RGB)
    img6 = ImageTk.PhotoImage(image=PIL.Image.fromarray(color_coverted))
    canvas2.create_image(300, 210, image=img6)
    canvas2.image = img6

def powerlaw():
    global img_path, img7, imgg
    img = Image.open(img_path)
    img.thumbnail((350, 350))
    imgUMat = img
    gray = cv2.cvtColor(np.float32(imgUMat), cv2.COLOR_RGB2GRAY)
    gray = np.uint8(imgUMat)
    imgg = np.array(255*(gray / 255) ** 2, dtype = 'uint8')
    imgg = np.array(imgg, dtype=np.uint8)
    color_coverted = cv2.cvtColor(imgg, cv2.COLOR_BGR2RGB)
    img7 = ImageTk.PhotoImage(image=PIL.Image.fromarray(color_coverted))
    canvas2.create_image(300, 210, image=img7)
    canvas2.image = img7

def canny(event=None):
    global img_path, img8, imgg
    img = Image.open(img_path)
    img.thumbnail((350, 350))
    imgUMat = img
    gray = cv2.cvtColor(np.float32(imgUMat), cv2.COLOR_RGB2GRAY)
    gray = np.uint8(imgUMat)
    imgg = cv.Canny(gray,100,200)
    imgg = np.array(imgg, dtype=np.uint8)
    color_coverted = cv2.cvtColor(imgg, cv2.COLOR_BGR2RGB)
    img8 = ImageTk.PhotoImage(image=PIL.Image.fromarray(color_coverted))
    canvas2.create_image(300, 210, image=img8)
    canvas2.image = img8
    
def laplacian(event=None):
    global img_path, img9, imgg
    img = Image.open(img_path)
    img.thumbnail((350, 350))
    imgUMat = img
    gray = cv2.cvtColor(np.float32(imgUMat), cv2.COLOR_RGB2GRAY)
    imgg = cv2.Laplacian(gray,cv2.CV_32F)
    imgg = np.array(imgg, dtype=np.uint8)
    color_coverted = cv2.cvtColor(imgg, cv2.COLOR_BGR2RGB)
    img9 = ImageTk.PhotoImage(image=PIL.Image.fromarray(color_coverted))
    canvas2.create_image(300, 210, image=img9)
    canvas2.image = img9
    
def sobelx(event=None):
    global img_path, img10, imgg
    img = Image.open(img_path)
    img.thumbnail((350, 350))
    imgUMat = img
    gray = cv2.cvtColor(np.float32(imgUMat), cv2.COLOR_RGB2GRAY)
    imgg = cv2.Sobel(gray,cv2.CV_64F,1,0,ksize=5)
    imgg = np.array(imgg, dtype=np.uint8)
    color_coverted = cv2.cvtColor(imgg, cv2.COLOR_BGR2RGB)
    img10 = ImageTk.PhotoImage(image=PIL.Image.fromarray(color_coverted))
    canvas2.create_image(300, 210, image=img10)
    canvas2.image = img10
    
def sobely(event=None):
    global img_path, img11, imgg
    img = Image.open(img_path)
    img.thumbnail((350, 350))
    imgUMat = img
    gray = cv2.cvtColor(np.float32(imgUMat), cv2.COLOR_RGB2GRAY)
    imgg = cv2.Sobel(gray,cv2.CV_64F,0,1,ksize=5)
    imgg = np.array(imgg, dtype=np.uint8)
    color_coverted = cv2.cvtColor(imgg, cv2.COLOR_BGR2RGB)
    img11 = ImageTk.PhotoImage(image=PIL.Image.fromarray(color_coverted))
    canvas2.create_image(300, 210, image=img11)
    canvas2.image = img11
    
def sobelxy(event=None):
    global img_path, img12, imgg
    img = Image.open(img_path)
    img.thumbnail((350, 350))
    imgUMat = img
    gray = cv2.cvtColor(np.float32(imgUMat), cv2.COLOR_RGB2GRAY)
    imgg = cv2.Sobel(gray,cv2.CV_64F,1,1,ksize=5)
    imgg = np.array(imgg, dtype=np.uint8)
    color_coverted = cv2.cvtColor(imgg, cv2.COLOR_BGR2RGB)
    img12 = ImageTk.PhotoImage(image=PIL.Image.fromarray(color_coverted))
    canvas2.create_image(300, 210, image=img12)
    canvas2.image = img12
    
def houghline():
    global img_path, img12, imgg
    img = Image.open(img_path)
    img.thumbnail((350, 350))
    imgUMat = img
    #gray = cv2.cvtColor(np.float32(imgUMat), cv2.COLOR_RGB2GRAY)
    #gray = np.uint8(imgUMat)
    gray = cv2.cvtColor(np.uint8(imgUMat), cv2.COLOR_RGB2GRAY)
    edges = cv.Canny(gray,50,150,apertureSize = 3)
    lines = cv.HoughLines(edges,1,np.pi/180,200)
    if lines is not None:
        for line in lines:
            rho,theta = line[0]
            a = np.cos(theta)
            b = np.sin(theta)
            x0 = a*rho
            y0 = b*rho
            x1 = int(x0 + 1000*(-b))
            y1 = int(y0 + 1000*(a))
            x2 = int(x0 - 1000*(-b))
            y2 = int(y0 - 1000*(a))
            cv.line(img,(x1,y1),(x2,y2),(0,0,255),2)
    imgg = np.array(img, dtype=np.uint8)
    color_coverted = cv2.cvtColor(imgg, cv2.COLOR_BGR2RGB)
    img13 = ImageTk.PhotoImage(image=PIL.Image.fromarray(color_coverted))
    canvas2.create_image(300, 210, image=img13)
    canvas2.image = img13
        
        
def houghcircle():
    global img_path, img14, imgg
    img = Image.open(img_path)
    img.thumbnail((350, 350))
    imgUMat = img
    gray = cv2.cvtColor(np.uint8(imgUMat), cv2.COLOR_RGB2GRAY)
    circles = cv.HoughCircles(gray,cv.HOUGH_GRADIENT,1,20,param1=50,param2=30,minRadius=0,maxRadius=0)
    circles = np.uint16(np.around(circles))
    for i in circles[0,:]:
        #cv.circle(img,(i[0],i[1]),i[2],(0,255,0),2)
        #cv.circle(img,(i[0],i[1]),2,(0,0,255),3)
        cv.circle(numpy.array(imgUMat),(i[0],i[1]),i[2],(0,255,0),2)
        cv.circle(numpy.array(imgUMat),(i[0],i[1]),2,(0,0,255),3)
    imgg = np.array(imgUMat, dtype=np.uint8)
    color_coverted = cv2.cvtColor(imgg, cv2.COLOR_BGR2RGB)
    img14 = ImageTk.PhotoImage(image=PIL.Image.fromarray(color_coverted))
    canvas2.create_image(300, 210, image=img14)
    canvas2.image = img14
    
def erosion(event=None):
    global img_path, img15, imgg
    img = Image.open(img_path)
    img.thumbnail((350, 350))
    imgUMat = img
    gray = cv2.cvtColor(np.float32(imgUMat), cv2.COLOR_RGB2GRAY)
    kernel = np.ones((5,5),np.uint8)
    imgg = cv2.erode(gray,kernel,iterations = 1)
    imgg = np.array(imgg, dtype=np.uint8)
    color_coverted = cv2.cvtColor(imgg, cv2.COLOR_BGR2RGB)
    img15 = ImageTk.PhotoImage(image=PIL.Image.fromarray(color_coverted))
    canvas2.create_image(300, 210, image=img15)
    canvas2.image = img15
    
def dilation(event=None):
    global img_path, img16, imgg
    img = Image.open(img_path)
    img.thumbnail((350, 350))
    imgUMat = img
    gray = cv2.cvtColor(np.float32(imgUMat), cv2.COLOR_RGB2GRAY)
    kernel = np.ones((5,5),np.uint8)
    imgg = cv2.dilate(gray,kernel,iterations = 1)
    imgg = np.array(imgg, dtype=np.uint8)
    color_coverted = cv2.cvtColor(imgg, cv2.COLOR_BGR2RGB)
    img16 = ImageTk.PhotoImage(image=PIL.Image.fromarray(color_coverted))
    canvas2.create_image(300, 210, image=img16)
    canvas2.image = img16

def opening(event=None):
    global img_path, img17, imgg
    img = Image.open(img_path)
    img.thumbnail((350, 350))
    imgUMat = img
    gray = cv2.cvtColor(np.float32(imgUMat), cv2.COLOR_RGB2GRAY)
    imgUMat = img
    gray = cv2.cvtColor(np.float32(imgUMat), cv2.COLOR_RGB2GRAY)
    kernel = np.ones((5,5),np.uint8)
    imgg = cv2.morphologyEx(gray, cv2.MORPH_OPEN, kernel)
    imgg = np.array(imgg, dtype=np.uint8)
    color_coverted = cv2.cvtColor(imgg, cv2.COLOR_BGR2RGB)
    img17 = ImageTk.PhotoImage(image=PIL.Image.fromarray(color_coverted))
    canvas2.create_image(300, 210, image=img17)
    canvas2.image = img17
    
def closing(event=None):
    global img_path, img18, imgg
    img = Image.open(img_path)
    img.thumbnail((350, 350))
    imgUMat = img
    gray = cv2.cvtColor(np.float32(imgUMat), cv2.COLOR_RGB2GRAY)
    imgUMat = img
    gray = cv2.cvtColor(np.float32(imgUMat), cv2.COLOR_RGB2GRAY)
    kernel = np.ones((5,5),np.uint8)
    imgg = cv2.morphologyEx(gray, cv2.MORPH_CLOSE, kernel)
    imgg = np.array(imgg, dtype=np.uint8)
    color_coverted = cv2.cvtColor(imgg, cv2.COLOR_BGR2RGB)
    img18 = ImageTk.PhotoImage(image=PIL.Image.fromarray(color_coverted))
    canvas2.create_image(300, 210, image=img18)
    canvas2.image = img18
    
def gradient(event=None):
    global img_path, img19, imgg
    img = Image.open(img_path)
    img.thumbnail((350, 350))
    imgUMat = img
    gray = cv2.cvtColor(np.float32(imgUMat), cv2.COLOR_RGB2GRAY)
    imgUMat = img
    gray = cv2.cvtColor(np.float32(imgUMat), cv2.COLOR_RGB2GRAY)
    kernel = np.ones((5,5),np.uint8)
    imgg = cv2.morphologyEx(gray, cv2.MORPH_GRADIENT, kernel)
    imgg = np.array(imgg, dtype=np.uint8)
    color_coverted = cv2.cvtColor(imgg, cv2.COLOR_BGR2RGB)
    img19 = ImageTk.PhotoImage(image=PIL.Image.fromarray(color_coverted))
    canvas2.create_image(300, 210, image=img19)
    canvas2.image = img19
    
def tophat(event=None):
    global img_path, img20, imgg
    img = Image.open(img_path)
    img.thumbnail((350, 350))
    imgUMat = img
    gray = cv2.cvtColor(np.float32(imgUMat), cv2.COLOR_RGB2GRAY)
    imgUMat = img
    gray = cv2.cvtColor(np.float32(imgUMat), cv2.COLOR_RGB2GRAY)
    kernel = np.ones((5,5),np.uint8)
    img = cv2.morphologyEx(gray, cv2.MORPH_TOPHAT, kernel)
    imgg = np.array(imgg, dtype=np.uint8)
    color_coverted = cv2.cvtColor(imgg, cv2.COLOR_BGR2RGB)
    img20 = ImageTk.PhotoImage(image=PIL.Image.fromarray(color_coverted))
    canvas2.create_image(300, 210, image=img20)
    canvas2.image = img20
    
def blackhat(event=None):
    global img_path, img21, imgg
    img = Image.open(img_path)
    img.thumbnail((350, 350))
    imgUMat = img
    gray = cv2.cvtColor(np.float32(imgUMat), cv2.COLOR_RGB2GRAY)
    imgUMat = img
    gray = cv2.cvtColor(np.float32(imgUMat), cv2.COLOR_RGB2GRAY)
    kernel = np.ones((5,5),np.uint8)
    imgg = cv2.morphologyEx(gray, cv2.MORPH_BLACKHAT, kernel)
    imgg = np.array(imgg, dtype=np.uint8)
    color_coverted = cv2.cvtColor(imgg, cv2.COLOR_BGR2RGB)
    img21 = ImageTk.PhotoImage(image=PIL.Image.fromarray(color_coverted))
    canvas2.create_image(300, 210, image=img21)
    canvas2.image = img21
    
def add_noise(event=None):
    global img_path, img22, imgg
    img = Image.open(img_path)
    img.thumbnail((350, 350))
    row = img.size[0]
    col = img.size[1]
    number_of_pixels = rand.randrange(300, 10000)
    for i in range(number_of_pixels):
        y_coord = rand.randrange(0, row - 1)
        x_coord = rand.randrange(0, col - 1)
        img.putpixel((y_coord, x_coord), 255)
    number_of_pixels = rand.randrange(300, 10000)
    for i in range(number_of_pixels):
        y_coord = rand.randrange(0, row - 1)
        x_coord = rand.randrange(0, col - 1)
        img.putpixel((y_coord, x_coord),0)
    imgg = np.array(img, dtype=np.uint8)
    color_coverted = cv2.cvtColor(imgg, cv2.COLOR_BGR2RGB)
    img22 = ImageTk.PhotoImage(image=PIL.Image.fromarray(color_coverted))
    canvas2.create_image(300, 210, image=img22)
    canvas2.image = img22

canvas2 = Canvas(root, width="600", height="420", relief=RIDGE, bd=2)
canvas2.place(x=15, y=15)
menubar = Menu(root)
file = Menu(menubar, tearoff=0)
menubar.add_cascade(label='File', menu=file)
file.add_command(label='Open', command=load)
file.add_command(label='Save', command=save)
file.add_command(label='Exit', command=root.destroy)
power = Menu(menubar, tearoff=0)
menubar.add_cascade(label='Transformation', menu=power)
power.add_command(label='Log Transform', command=logtransform)
power.add_command(label='Power-Law', command=powerlaw)
hough = Menu(menubar, tearoff=0)
menubar.add_cascade(label='Hough', menu=hough)
hough.add_command(label='Line', command=houghline)
hough.add_command(label='Circle', command=houghcircle)
filter = Menu(menubar, tearoff=0)
menubar.add_cascade(label='Filtering', menu=filter)
filter.add_command(label='Pepper and Salt', command=add_noise)
filter.add_command(label='Blurring', command=blur)
filter.add_command(label='BoxBlurring', command=boxblur)
filter.add_command(label='contour', command=contour)
filter.add_command(label='Emboss', command=emboss)
filter.add_command(label='Edge Enhance', command=edgeenhance)
edge = Menu(menubar, tearoff=0)
menubar.add_cascade(label='Edge Detection', menu=edge)
edge.add_command(label='Laplacian', command=laplacian)
edge.add_command(label='Canny', command=canny)
edge.add_command(label='Sobel X', command=sobelx)
edge.add_command(label='Sobel Y', command=sobely)
edge.add_command(label='Sobel XY', command=sobelxy)
hough = Menu(menubar, tearoff=0)
morphology = Menu(menubar, tearoff=0)
menubar.add_cascade(label='Morphology', menu=morphology)
morphology.add_command(label='Erosion', command=erosion)
morphology.add_command(label='Dilation', command=dilation)
morphology.add_command(label='Opening', command=opening)
morphology.add_command(label='Closing', command=closing)
morphology.add_command(label=' Morphological Gradient', command=gradient)
morphology.add_command(label='Top Hat', command=tophat)
morphology.add_command(label='Black Hat', command=blackhat)
root.config(menu=menubar)
root.mainloop()
```
